```python
import numpy as np
from matplotlib.pyplot import subplots
rng = np.random.default_rng(1003)
x = rng.standard_normal(size = 100)
y = rng.standard_normal(size = 100)
fig, ax = subplots(figsize=(8, 8))
ax.plot(x, y)
```

```python
fig, ax = subplots(figsize=(8, 8))
ax.plot(x, y, 'o')
```

### The scatter function is more powerful than the plot function when it comes to creating scatter graphic

```python
fig, ax = subplots(figsize = (10,8))
ax.scatter(x, y, marker = 'o')
ax.set_xlabel("X lable")
ax.set_ylabel("Y lable")
ax.set_title("Test")
```

```python
fig, axes = subplots(nrows = 2, ncols = 3, figsize = (15, 5))
```

```python
axes[0,1].plot(x, y, 'o')
axes[1,2].scatter(x, y, marker = 'x')
fig
```

```python
fig.savefig("testPhotos/Test1.png", dpi = 200)
```

```python
axes[0,1].set_xlim(-3,3)
axes[0,1].set_ylim(-3,3)
fig
```
